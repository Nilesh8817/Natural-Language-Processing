```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation
from nltk.stem import WordNetLemmatizer

def text_preprocessing(document):

  # tokenization & lower case
  tokens = word_tokenize(document.lower())

  # stopword and punctuation removal
  stop = stopwords.words("english")
  punc = list(punctuation)
  bad_tokens = stop + punc

  clean_tokens = [token for token in tokens if token not in bad_tokens]

  #Lemmatization
  lemmatizer = WordNetLemmatizer()
  lemmatized_tokens = [lemmatizer.lemmatize(token) for token in clean_tokens]

  return " ".join(lemmatized_tokens)
```

```python
text_preprocessing(document)
```

## Vectorization

```python
import pandas as pd
```

```python
d1 = "I am facing network issue, Please solve the network problem"
d2 = "I am unable to make phone calls"
d3 = "I am shifting, kindly port my number"
d4 = "I want to port my number"
```

```python
corpus = pd.Series([d1,d2,d3,d4])
```

```python
corpus
```

### CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#cv = CountVectorizer()
cv = CountVectorizer(stop_words="english")
```

```python
corpus_count = cv.fit_transform(corpus)
```

```python
corpus_count
```

```python
# optional - to see how our data looks like
corpus_count.toarray()
```

```python
cv.get_feature_names()
```

```python
pd.DataFrame(corpus_count.toarray(), columns=cv.get_feature_names())
```

### Tfidf vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#tfidf = TfidfVectorizer()
tfidf = TfidfVectorizer(stop_words="english")
```

```python
corpus_tfidf = tfidf.fit_transform(corpus)
```

```python
corpus_tfidf
```

```python
# optional
corpus_tfidf.toarray()
```

```python
tfidf.get_feature_names()
```
